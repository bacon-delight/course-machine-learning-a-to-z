# Decision Tree Regression

Consider the following example of Decision Tree Regression. We try to do multiple splits on the dataset with respect to two coordinate axes and create a decision tree based on the averages of each split.

### Splits with averages on each split


### Decision Tree


```python
# Importing the Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing Dataset
dataset = pd.read_csv('./Position_Salaries.csv')
dataset.head()
```

```python
# Preparing the Dataset
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
# Visualizing the new arrays
print("X = \n", X)
print("y = \n", y)
```

Since we have a very small dataset, we'll use the entire dataset for training the model and avoid splitting it.

```python
# Fitting Decision Tree Regression to the Dataset
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X, y)
```

```python
# Predicting a new result
print(regressor.predict([[6.5]]))
```

### Visualizations

```python
# Visualising the Regression results (High Resolution)
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Regression results (Low Resolution)
plt.scatter(X, y, color = 'red', marker="x")
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

### Applying Decision Tree in a Large Dataset
- Take care of missing values (Imputer)
- Resolve categorial data (Encode)
- Feature scaling is not needed in Decision Tree
- If dataset is big enough, split into training and test set
- If a single observation has several features, prediction will be like `regressor.predict([feature1, feature2, ...])`
- The visualizations above will not work for datasets with multiple features
